```python
import numpy as np
from numpy import random
import pandas as pd
```

```python
results = []
for n in range (1,100):
    product = []
    for i in range (1000):
        variances = []
        a = random.rand(1,n)
        b = random.rand(n,1)
        c = np.dot(a,b)
        product.append(c[0][0])
    var= np.var(np.array(product))
    variances.append(var)
    results.append({'dimension': n, 'variance': var})
```

```python
 final = pd.DataFrame(results)
```

```python
final['scaled'] = final['variance']/np.sqrt(final.dimension)
```

```python

import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))

plt.plot(final.dimension,final.scaled,label= 'scaled variance divided by sqrt of dimension')
plt.plot(final.dimension,final.variance,label = 'simple variance')

plt.xlabel('dimension')
plt.ylabel('Variance')
plt.title('variance and scaled variance vs vector dimension')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
